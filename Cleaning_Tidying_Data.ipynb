```
weather  = pd.read_csv('/content/sample_data/weather.csv')
weather_melt = weather.melt(id_vars= ['id','year','month','element'], var_name= 'day', value_name= 'temp')
weather_tidy = weather_melt.pivot_table(
                                    index = ['id','year','month','day'],
                                    columns = 'element',
                                    values = 'temp'
)

weather_tidy.reset_index()
```

```
billboard = pd.read_csv('/content/sample_data/billboard.csv')
billboard_melt = pd.melt(billboard, id_vars= ['year',	'artist','track'	,'time',	'date.entered'],
                         var_name='week',
                         value_name='rating')
#billboard_melt.loc[(billboard_melt['year']==2000) & (billboard_melt['artist']=='2 Pac'),:]
billboard_song_columns = ['year', 'artist',	'track'	, 'time']
billboard_song = billboard_melt[billboard_song_columns]
billboard_song =billboard_song.drop_duplicates()
billboard_song['id'] =range(len(billboard_song)) 
# billboard_song.to_csv('billboard_song.csv', index= False)
billboard_rating =  billboard_melt.merge(billboard_song,
                                         how= 'inner',
                                         on = billboard_song_columns
)

#billboard_rating.loc[(billboard_melt['year']==2000) & (billboard_melt['artist']=='2 Pac'),:]

billboard_rating = billboard_rating [['id','date.entered','week' , 'rating']]
billboard_rating.to_csv('billboard_rating.csv', index= False)
```
